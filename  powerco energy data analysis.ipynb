```python
price_features = price_data.groupby('id').mean(numeric_only=True)

price_features.head()
```

```python
merged_df = client_data.merge(
    price_features,
    on='id',
    how='left'
)

merged_df.shape
```

```python
merged_df.groupby('churn')[
    'price_off_peak_var'
].mean()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(
    x='churn',
    y='price_off_peak_var',
    data=merged_df
)

plt.title("Price vs Churn")
plt.show()
```
